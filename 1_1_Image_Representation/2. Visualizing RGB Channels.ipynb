# RGB colorspace

### Import resources

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

### Read in an image

```python
# Read in the image
image = mpimg.imread('images/wa_state_highway.jpg')

plt.imshow(image)
```

### RGB channels 

Visualize the levels of each color channel. Pay close attention to the traffic signs!

```python
# Isolate RGB channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

# Visualize the individual color channels in gray scale
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r, cmap='gray')
ax2.set_title('G channel')
ax2.imshow(g, cmap='gray')
ax3.set_title('B channel')
ax3.imshow(b, cmap='gray')

## Which area has the lowest value for red? What about for blue?
```

```python
# Visualize individual color channels
f,(ax1,ax2,ax3)= plt.subplots(1,3,figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r)
ax2.set_title('G channel')
ax2.imshow(g)
ax3.set_title('B channel')
ax3.imshow(b)

```

```python
additive_image = r+g+b
plt.imshow(additive_image)
```

```python
import numpy as np
additive_axis= r
additive_axis= np.concatenate((additive_axis,g),axis=1)
plt.imshow(additive_axis)
```

```python
additive_axis= r
additive_axis= np.concatenate((additive_axis,g),axis=0)
plt.imshow(additive_axis)
```

```python
additive_axis= np.dstack((r,g,b))
plt.imshow(additive_axis)
```

```python
#Find the maximum and minimum grayscale values in this image

max_val = np.amax(image)
min_val = np.amin(image)

print('Max: ', max_val)
print('Min: ', min_val)
```
